## Plot accuracy/layers

```python
def plot_accuracy_comparison(layers, supernet_data, scratch_data, title):
    sns.set_style("whitegrid")
    sns.set_palette(["#000000", "#00008B"])

    plt.figure(figsize=(14, 8))

    ax1 = plt.gca()
    ax1.plot(
        layers,
        supernet_data,
        marker="o",
        linestyle="--",
        label="SuperNet",
        color=sns.color_palette()[0],
    )

    ax2 = ax1.twinx()
    ax2.plot(
        layers,
        scratch_data,
        marker="o",
        label="From scratch",
        color=sns.color_palette()[1],
    )

    ax1.set_xlabel("Layers", fontsize=14)
    ax1.set_ylabel(
        "Top-1 accuracy for SuperNet", fontsize=14, color=sns.color_palette()[0]
    )
    ax2.set_ylabel(
        "Top-1 accuracy for From Scratch", fontsize=14, color=sns.color_palette()[1]
    )
    plt.title(title, fontsize=16)
    ax1.set_xticks(layers)

    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines1 + lines2, labels1 + labels2, loc=0)

    plt.tight_layout()
    plt.show()
```

```python
plot_accuracy_comparison(
    results_df["Layers"],
    results_df["SuperNet_32_64"],
    results_df["From scratch 32_64"],
    "Comparison of Top-1 accuracy for 32_64 configuration",
)
```
